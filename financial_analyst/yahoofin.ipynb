```python
# Debt, Equity, Beta, tax rate, market return
beta = ticker.info['beta']
tax_rate_us = 0.15
market_return = 0.10
totaldebt = ticker.info['totalDebt']
equity = totaldebt / ticker.info['debtToEquity']
equity
marketcap = ticker.info['marketCap']
equity
```

```python
ticker.info['debtToEquity']
```

```python
# Ten Year Treasury Rate
risk_free_rate = yf.Ticker('^TNX').info['previousClose']/100
risk_free_rate
```

```python
#cost of equity

cost_of_equity = risk_free_rate + beta * (market_return - risk_free_rate)
cost_of_equity
```

```python
#cost of debt
cost_of_debt = 0.05
```

```python
#wacc = ((equity/equity+totaldebt) * cost_of_equity) + ((totaldebt/equity+totaldebt) * cost_of_debt) * (1-tax_rate_us)
wacc = (equity/(equity+totaldebt))*cost_of_equity + (totaldebt / (equity+totaldebt)) * cost_of_debt * (1-tax_rate_us)

wacc
```

```python
cashflow = ticker.cashflow

cashflow
```

```python
# Free Cash Flow Estimate for future

growth_rate = 0.02
ev_to_ebitda = ticker.info['enterpriseToEbitda']
free_cash_flow = ticker.info['freeCashflow']
terminal_period = 5
free_cash_flow
teminal_value_fcf = free_cash_flow*(1+growth_rate)**terminal_period

#cashflow.index.tolist()
#len(cashflow.columns)
ending_fcf = cashflow.loc['Free Cash Flow'].iloc[0]
beginning_fcf = cashflow.loc['Free Cash Flow'].iloc[-1]

terminal_period = 4
while pd.isna(beginning_fcf):
    beginning_fcf = cashflow.loc['Free Cash Flow'].iloc[terminal_period-1]
    terminal_period -= 1

beginning_fcf

fcf_cagr = (ending_fcf/beginning_fcf)**(1/terminal_period) - 1
fcf_cagr

future_fcf = ending_fcf*(1+fcf_cagr)**(5)
future_fcf
```

```python
# income statement

income_statement = ticker.income_stmt
income_statement
balance_sheet = ticker.balance_sheet
balance_sheet
```

```python
#Ebitda forecast

ending_ebitda = income_statement.loc['EBITDA'].iloc[0]
beginning_ebitda = income_statement.loc['EBITDA'].iloc[len(income_statement.columns)-1]
beginning_ebitda

ebitda_cagr = (ending_ebitda/beginning_ebitda)**(1/len(income_statement.columns)) - 1
ebitda_cagr
```

```python
future_ebitda = ending_ebitda*(1+ebitda_cagr)**(5)
future_ebitda
```

```python
# Terminal Value - Perpetuity Growth

pg_tv = future_fcf*(1+growth_rate)/(wacc - growth_rate)
pg_tv
```

```python
# Terminal Value - ebitda multiple

eb_tv = future_ebitda * ev_to_ebitda
eb_tv
```

```python
pg_tv - eb_tv
```

```python
# Present Value of FCF, TV and Enterprise Value

projected_cash_flows = []
for year in range(1, 5):
    projected_fcf = ending_fcf*(1+fcf_cagr)**year
    projected_cash_flows.append(projected_fcf)

pv_fcf = 0
for year, fcf in enumerate(projected_cash_flows, start = 1):
    pv = fcf / (1+wacc)**(year+1)
    pv_fcf += pv

tv = (projected_cash_flows[-1]*(1+growth_rate))/(wacc - growth_rate)
tv

pv_tv = tv/((1+growth_rate)**5)
pv_tv

ev = pv_fcf+pv_tv
ev
```

```python
#Calculate Equity

balance_sheet = ticker.balance_sheet
balance_sheet
cash_and_cash_equivalents = balance_sheet.loc['Cash And Cash Equivalents'].iloc[0]
cash_and_cash_equivalents

total_debt = balance_sheet.loc['Total Debt'].iloc[0]
total_debt

equity_value = ev + cash_and_cash_equivalents - total_debt
equity_value
```

```python
# Shares Outstanding
shares_outstanding = ticker.info['sharesOutstanding']
shares_outstanding

share_price = equity_value / shares_outstanding
share_price
```
